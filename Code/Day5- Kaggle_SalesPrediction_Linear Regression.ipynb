```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
data = pd.read_csv("Advertising.csv")
```

```python
data.head(10)
```

```python
data.columns
```

```python
data.drop('Unnamed: 0', axis =1)
```

```python
data.isnull().sum()
```

```python
py.figure(figsize=(10,8))
py.scatter(data['TV'],data['sales'],c='Black')
py.xlabel("Money spent on TV ads")
py.ylabel("sales")
py.show()
```

```python
py.figure(figsize=(10,8))
py.scatter(data['radio'],data['sales'],c='Black')
py.xlabel("Money spent on TV ads")
py.ylabel("sales")
py.show()
```

```python
py.figure(figsize=(10,8))
py.scatter(data['newspaper'],data['sales'],c='Black')
py.xlabel("Money spent on TV ads")
py.ylabel("sales")
py.show()
```

```python
X= data['TV'].values.reshape(-1,1)
Y = data['sales'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, Y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
py.figure(figsize=(16, 8))
py.scatter(
    data['TV'],
    data['sales'],
    c='black'
)
py.plot(
    data['TV'],
    predictions,
    c='blue',
    linewidth=2
)
py.xlabel("Money spent on TV ads ($)")
py.ylabel("Sales ($)")
py.show()
```

```python
#Testing the accuracy of model
X = data['TV']
y = data['sales']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
#R-squared is 0.612 which is close to 1 . Hence assuring 60% accurate prediction rate
#p-value is 0 and should be less than 0.05. Showing strong correlation between two values
```
