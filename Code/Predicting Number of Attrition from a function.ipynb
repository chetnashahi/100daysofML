```python
def f(row):
    if row>=0 and row<=1:
        t4='0-1'
    elif row>1 and row<=2:
        t4='1-2'
    elif row>2 and row<=3:
        t4='1-1.5'
    elif row>3 and row<=4:
        t4='3-4'
    elif row>4 and row<=5:
        t4='4-5'
    elif row>5 and row<=6:
        t4='5-6'
    else:
        t4=0
    return t4
```

```python
def f(row):
    if row>=0 and row<=0.5:
        t4='0-0.5'
    elif row>0.5 and row<=1:
        t4='0.5-1'
    elif row>1 and row<=1.5:
        t4='1-1.5'
    elif row>1.5 and row<=2:
        t4='1.5-2'
    elif row>2 and row<=2.5:
        t4='2-2.5'
    elif row>2.5 and row<=3:
        t4='2.5-3'
    elif row>3 and row<=3.5:
        t4='3-3.5'
    elif row>3.5 and row<=4:
        t4='3.5-4'
    elif row>4.5 and row<=5:
        t4='4.5-5'
    elif row>5 and row<=5.5:
        t4='5-5.5'
    elif row>5.5 and row<=6:
        t4='5.5-6'
    elif row>6 and row<=6.5:
        t4='6-6.5'
    elif row>6.5 and row<=7:
        t4='6.5-7'
    elif row>7 and row<=7.5:
        t4='7-7.5'
    elif row>7.5 and row<=8:
        t4='7.5-8'
    else:
        t4='0'
    return t4
```

```python
df['Time_Role_Bucket'] = df['TIME_IN_ROLE'].apply(lambda x: f(x))
```

```python
Total=dict(df['Time_Role_Bucket'].value_counts())
print(dict(Total))
```

```python
Terminated=dict(df2['Time_Role_Bucket'].value_counts())
print(dict(Terminated))
```

```python
list_2019={}
#print(len(Total.items()))
sharedKeys = set(Total.keys()).intersection(Terminated.keys())
for key in sharedKeys:
    #print('Key: {}, Total: {}, Terminated: {}, Probability: {}'.format(key, Total[key], Terminated[key],Terminated[key]/Total[key]))
    list_2019.update({key:(Terminated[key]/Total[key])})

print(dict(list_2019))
   
```
